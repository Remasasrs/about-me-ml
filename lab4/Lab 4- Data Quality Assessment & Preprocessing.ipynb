```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Amount'])

plt.title('Boxplt of Amount')
plt.show()
```

```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Amount'] < lower) | (df['Amount'] > upper)]
outliers.head(15)
```

```python
df_no_outliers = df[(df['Amount'] >= lower) & (df['Amount'] <= upper)]

print('df with outliers: ',df.shape)
print('df without outliers: ',df_no_outliers.shape)
```

```python
lower_cap = df['Amount'].quantile(0.05) # 5%
upper_cap = df['Amount'].quantile(0.95) # 95%

df_capped = df.copy()
df_capped['Amount'] = df_capped['Amount'].clip(lower_cap, upper_cap)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_capped['Amount'])

plt.title('Boxplt of Amount')
plt.show()
```

```python
df[['Amount','Time']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df[['Amount', 'Time']].copy()

df_scaled[['Amount', 'Time']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df[['Amount', 'Time']].copy()

df_standardized[['Amount', 'Time']] = scaler.fit_transform(df_standardized)

df_standardized.head()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df_standardized[['Amount','Time']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Before PCA)")
plt.show()
```

```python
from sklearn.decomposition import PCA

X = df_standardized[['Amount', 'Time']].dropna()

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
X = df_standardized[['Amount', 'Time']].dropna()

# Align Class with X
y = df.loc[X.index, 'Class']

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

# Plot
plt.figure(figsize=(6,4))
plt.scatter(principal_components[:,0], principal_components[:,1],
            c=y, cmap='coolwarm', alpha=0.5)
plt.title("PCA Projection (Fraud vs Normal)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
